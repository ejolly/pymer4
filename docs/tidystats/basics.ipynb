# Basic Usage

`tidystats` is an experimental new module in `pymer4` that aims to bring more R "tidy" functionality to Python and expand the scope of great libraries available. This demo includese 


```python
import pymer4.tidystats as ts
```

## Included datasets


```python
mtcars = ts.load_dataset('mtcars')
mtcars.head()
```

## Cleaning with janitor

```python
ts.get_dupes(mtcars, 'mpg', 'hp')
```

```python
titanic = ts.load_dataset('titanic_train').drop_nulls()
titanic.glimpse()
```

```python
titanic = ts.clean_names(titanic)
titanic.glimpse()
```

## Linear regression with lm

```python
df = ts.load_dataset('sleep')
model = ts.lm('Reaction ~ Days', data=df)
```

Basic functions that work with models like `resid`

```python
# First 3 residuals
ts.resid(model)[:3]
```

And `coef` 

```python
# Coefficients
ts.coef(model)
```

## Tidy model manipulation

`tidy`, `glance`, and `augment` from [broom](https://broom.tidymodels.org/)

```python
# Summary table
ts.tidy(model)
```

```python
# Fit statistics
ts.glance(model)
```

```python
# Augment original data with columns of model output
ts.augment(model).head()
```

## GLM and contrasts

String variables are encoded as factors by default:

```python
# sex is a string
model = ts.glm('survived ~ 1 + sex', data=titanic, family='binomial')

# Summary
ts.tidy(model)
```

We can confirm this using the `contrasts` argument to an `lm()`, `glm()`, `lmer()`, and `glmer()` model. This takes a Python dictionary with keys corresponding to variable names and values provides as a either a string referring to a contrast type in R.

```python
# contr.treatment is the default (dummy-coding)
model = ts.glm('survived ~ 1 + sex', data=titanic, family='binomial', contrasts = {'sex': 'contr.treatment'})

# Summary
ts.tidy(model)
```

## Categorical coding schemes

```python
poker = ts.load_dataset('poker')
poker.glimpse()
```

### Treatment/Dummy coding

```python
# Default dummy contrasts
model = ts.lm('balance ~ hand', data=poker)

# Inspect model design matrix
ts.model_matrix(model)
```

### Deviation/Sum coding

```python
# Sum-to-zero
sum_model = ts.lm('balance ~ hand', data=poker, contrasts={'hand': 'contr.sum'})

ts.model_matrix(sum_model)
```

### Orthogonal Polynomial coding

```python
# Poly
poly_model = ts.lm('balance ~ hand', data=poker, contrasts={'hand': 'contr.poly'})

ts.model_matrix(poly_model)
```

## Emmeans and Type-3 ANOVA

We can use the `joint_tests()` function from `emmeans()` to automatically perform the appropriate ANOVA style F-tests

```python
model = ts.lm('balance ~ hand * skill', data=poker)

# Handles encoding and type III calculation
ts.joint_tests(model)
```

## Emmeans marginal comparisons

We can also use `emmeans` to calculate the marginal means, aka "cell-means" from our ANOVA

```python
# Marginal means
ts.emmeans(model, specs='hand', by='skill')
```

`tidystats` exposes `emmeans::contrast` as the `contrasts` keyword argument for simplicity. Internally it combines the `emmeans()` and the `contrast()` functions as recommended

```python
# Pairwise comparisons
ts.emmeans(model, specs='hand', by='skill', contrasts='pairwise')
```

Which makes it very intuitive to specify custom contrasts and correction factors:

```python
# Custom contrasts
# Across levels of hand, marginalizing over skill
ts.emmeans(model, specs='hand', 
contrasts={
    'bad_neutral_good': [-1, 1, 0],
    '2bad_others': [1, -.5, -.5]
    }, 
adjust='bonferroni')
```
