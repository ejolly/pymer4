# Piping in `tidystats` like `magittr`

`tidystats` makes use of the great `toolz` library to offer basic piping functionality. Just using `pipe()` and for each step of your pipe separate your function and its arguments into `ts.do()` as separate arguments

```python
import tidystats as ts

df = ts.load_dataset()

ts.pipe(df,
    ts.do(ts.lm, 'Reaction ~ Days'),
    ts.do(ts.augment),
    ts.do(lambda df: df.head())
)
```
