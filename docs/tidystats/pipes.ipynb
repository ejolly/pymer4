`tidystats` offers a piping API similar to `magittr` in R using `>>`

Importing from `pymer4.tidystats.pipe` instead of `pymer4.tidystats` will enable this functionality.  
Numerous `polars` DataFrame methods are also available for use in pipes

```python
import pymer4.tidystats.pipe as ts

from pymer4.tidystats import load_dataset
from polars import col
sleep = load_dataset('sleep')
```

```python
# Fit a regression and get the first 5 predicted values
(
    sleep
    >> ts.lm("Reaction ~ Days")
    >> ts.augment()
    >> ts.select("Reaction", "fitted")
    >> ts.head()
)
```

```python
# Inspect the model fit for a regression using data from days > 1 only
(
    sleep
    >> ts.filter(col('Days') > 1)
    >> ts.lm('Days ~ 1')
    >> ts.glance()
    >> ts.select('r_squared', 'AIC', 'logLik')
)
```

```python
# Fit an LMM and look at the fixed effect estimate only
(
    sleep
    >> ts.lmer('Days ~ 1 + (1 | Subject)')
    >> ts.tidy()
    >> ts.filter(col('effect') == 'fixed')
)
```
